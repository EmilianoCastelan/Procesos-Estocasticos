**¿Qué es la caminta aleatoria?**

 Es una formalización matemática de la trayectoria que resulta de hacer sucesivos pasos aleatorios.

 Para este ejercicio consideraremos una caminata aleatoria unidimensional definida por:

  $x_n=∑_{i=1}^{n}S_i$

donde cada paso $S_i$ es una variable aleatoria tal que:

$Sᵢ =
\begin{cases}
+1 & \text{con probabilidad } p \\
-1 & \text{con probabilidad } q
\end{cases}$

Y que ademas cumpla $p+q=1$



```python
import random as ran
import matplotlib.pyplot as plt
```

```python
"""
n= número de pasos
P= probabilidad de ir a la derecha
x0= posición inicial
 """

def caminata_1D(n,p,x0=0):
  posicion=x0
  trayectoria=[posicion]

  for i in range(n):
    r=ran.random()
    if r < p:
        paso = 1
    else:
        paso = -1

    posicion += paso
    trayectoria.append(posicion)

  return trayectoria
```

```python
def graficar_caminata(n=200, p=0.5):
    trayectoria = caminata_1D(n, p)

    plt.figure(figsize=(8,5))
    plt.plot(trayectoria)
    plt.axhline(0, color='black')
    plt.grid(True)
    plt.title(f"Caminata Aleatoria (p={p})")
    plt.xlabel("Paso")
    plt.ylabel("Posición")
    plt.show()


graficar_caminata(n=200, p=0.5)
```
